```python
#QQ Plot
# A Q-Q plot (Quantile-Quantile plot) is used to check whether a dataset follows a normal distribution.
# It plots the quantiles of the dataset against the theoretical quantiles of
# a normal distribution. If the points in the plot roughly follow a straight diagonal line, the data is normally distributed.

import scipy.stats as stats
import matplotlib.pyplot as plt
import statsmodels.api as sm

unit1 = df["Unit A"]
unit2 = df["Unit B"]

sm.qqplot(unit1, line='s')
plt.title("Q-Q Plot for Unit A")
plt.show()

sm.qqplot(unit2, line='s')
plt.title("Q-Q Plot for Unit B")
plt.show()
```

```python
#nomal distrubion 
```

```python
nit1 = df["Unit A"]
unit2 = df["Unit B"]

# Check normality using Shapiro-Wilk test
print("Shapiro Test for Unit 1:", stats.shapiro(unit1))
print("Shapiro Test for Unit 2:", stats.shapiro(unit2))
```

```python
print("Levene‚Äôs Test:", stats.levene(unit1, unit2))
t_stat, p_value = stats.ttest_ind(unit1, unit2, equal_var=True)  # Use equal_var=False if variances are unequal
```

```python
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in cutlet diameter.")
else:
    print("Fail to reject the null hypothesis: No significant difference in cutlet diameter.")
```
