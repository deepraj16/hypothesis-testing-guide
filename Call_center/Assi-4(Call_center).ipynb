# Best Statistical Test: Chi-Square Test for Independence


Since the data consists of categorical variables (center and defect status), a Chi-Square Test for Independence is the appropriate test.

```python
import pandas as pd 
import numpy as np 
df= pd.read_csv("CustomerOrderform.csv")
```

```python
df.sample(3)
```

```python
import matplotlib.pyplot as plt 
ax=plt.figure(figsize=(12,4))
plt.subplot(1,4,1)
df['Phillippines'].value_counts().plot(kind='pie')

plt.subplot(1,4,2)
df['India'].value_counts().plot(kind='pie')

plt.subplot(1,4,3)
df['Malta'].value_counts().plot(kind='pie')

plt.subplot(1,4,4)
df['Indonesia'].value_counts().plot(kind='pie')
```

```python
# Count the defective and non-defective forms for each center
contingency_table = df.apply(pd.Series.value_counts).T
print("Contingency Table:\n", contingency_table)
```

# perfroming chic square test 

```python
import scipy.stats as stats
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

# Print results
print("Chi-Square Statistic:", chi2_stat)
print("p-value:", p_value)

# Conclusion
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The defective % varies across centers.")
else:
    print("Fail to reject the null hypothesis: No significant difference in defective % among centers.")
```
