```python
lab1 = df["Laboratory_1"]
lab2 = df["Laboratory_2"]
lab3 = df["Laboratory_3"]
lab4 = df["Laboratory_4"]
```

```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
# Check normality using Shapiro-Wilk Test
print("Shapiro Test for Laboratory 1:", stats.shapiro(lab1))
print("Shapiro Test for Laboratory 2:", stats.shapiro(lab2))
print("Shapiro Test for Laboratory 3:", stats.shapiro(lab3))
print("Shapiro Test for Laboratory 4:", stats.shapiro(lab4))
```

```python
print("Levene’s Test:", stats.levene(lab1, lab2, lab3, lab4))

# Perform One-Way ANOVA
f_statistic, p_value = stats.f_oneway(lab1, lab2, lab3, lab4)

# Print results
print("F-statistic:", f_statistic)
print("p-value:", p_value)
```

```python
# Conclusion
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in average TAT among the laboratories.")
else:
    print("Fail to reject the null hypothesis: No significant difference in average TAT among the laboratories.")
```
